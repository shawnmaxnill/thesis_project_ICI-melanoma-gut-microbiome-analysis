```python
# Plotting function
def plot_dot(df):
    
    # Formatting
    df_formatted = df.stack().reset_index()
    df_formatted.columns = ["row", "column", "value"]
    
    # Plot
    plt.figure(figsize=(14, 6))
    sns.stripplot(data = df_formatted, x = "column", y = "value")
    plt.title("Absolute Difference Across Samples")
    plt.xlabel("Samples")
    plt.ylabel("Absolute Difference Value")
    plt.xticks(rotation=65, ha = "right", fontsize=8)
    plt.tight_layout()
    plt.show()
    
plot_dot(diff_sup_df), plot_dot(diff_fece_df), plot_dot(FF107_diff_df)
```

## Sample duplicate handling and in-depth breakdown (WIP)

It seems that sample `179fece00032` has two same column names but with different values, we will also dive into that.

```python
# Storing unique samples based on the plotted graph
diff_sup_samples = ["179supB00069", "179supB00610"]
diff_fece_samples = ["179fece00015", "179fece00022", "179fece000232"]
quads = ["179fece00032"]

# Dataframe creator
def making_df(df, sample_name, main_df):

    return pd.DataFrame({"taxon_name" : df[sample_name].index,
                         "difference" : df[sample_name].values,
                         "duplicate_a" : main_df[0][sample_name].values,
                         "duplicate_b" : main_df[1][sample_name].values}).sort_values(by = "difference", ascending = False)
```

```python
# For 'sup' samples 
sup142_df = [making_df(diff_sup_df, i, supernatant107_df_lst) for i in diff_sup_samples] # 2 samples

# For 'fece' samples
fece107_df = [making_df(diff_fece_df, i, fece142_df_lst) for i in diff_fece_samples] # 3 samples
```

```python
# Viewing duplicates for sup142_df "179supB00069", "179supB00610"
for df in sup142_df:
    display(df.iloc[:20,:])
    
# Viewing duplicates for fece107_df "179fece00015", "179fece00022", "179fece000232"
for df in fece107_df:
    display(df.iloc[:20,:])
    
# Viewing duplicates for FF samples
FF_merged = pd.concat([FF107_samples_df[0], FF107_samples_df[1]], axis = 1)
FF_merged = FF_merged.reset_index(drop = True)
FF_merged = FF_merged[sorted(FF_merged.columns)]
FF_merged.iloc[:20,:]
```

Lets plot them for more visual clarity

```python
# plotting function
def plot_comparison(df):
        
    # Allocating index for each value
    x = range(len(df))

    # Plot config
    plt.figure(figsize=(10, 6))
    plt.scatter(x, df["duplicate_a"], label="duplicate_a", color="blue", alpha = 0.4)
    plt.scatter(x, df["duplicate_b"], label="duplicate_b", color="orange", alpha = 0.4)

    # Details in plot
    plt.xlabel("Taxa Index")
    plt.ylabel("Abundance")
    plt.yticks(range(0, 101, 5))  # From 1 to 75, stepping by 5
    plt.title("Dot Plot: Differences between duplicate_a and duplicate_b")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
for df in sup142_df: # "179supB00069", "179supB00610"
    plot_comparison(df)
```

```python
for df in fece107_df: # "179fece00015", "179fece00022", "179fece000232"
    plot_comparison(df)
```

```python
# Plotting function for FF_samples
def plot_comparison2(df, colname1, colname2):
        
    # Allocating index for each value
    x = range(len(df))

    # Plot config
    plt.figure(figsize=(10, 6))
    plt.scatter(x, df[colname1], label = colname1, color = "blue", alpha = 0.5)
    plt.scatter(x, df[colname2], label = colname2, color = "red", alpha = 0.3)

    # Details in plot
    plt.xlabel("Taxa Index")
    plt.ylabel("Abundance")
    plt.yticks(range(0, 51, 5))  # From 1 to 75, stepping by 5
    plt.title("Dot Plot: Differences between duplicate_a and duplicate_b")
    plt.legend()
    plt.tight_layout()
    plt.show()

for i in range(0, len(FF_merged.columns), 2):
    
    col1 = FF_merged.columns[i]
    col2 = FF_merged.columns[i + 1]
    
    plot_comparison2(FF_merged, col1, col2)
```
